```python
class Kiosk:    
    def __init__(self):
        menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.menu = menu
        self.price = price
        self.order_menu = []
        self.order_price = []

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        print()
        n = -1
        while n != 0:
            n = int(input("음료 번호를 입력하세요 (0 입력 시 주문 완료): "))

            if n == 0:
                print("주문을 완료합니다.\n")
                for i in range(len(self.order_menu)):
                    print(self.order_menu[i], ":", self.order_price[i], "원")
                break
            
            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n and n <= len(self.menu):
                self.price_sum = self.price[n-1]

                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT" 
                    elif t == 2:
                        self.temp = "ICE"  
                    else:    
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_price.append(self.price[n-1])
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                print(self.temp, self.menu[n-1], "추가되었습니다.\n")

            else:  
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    #주문 취소 메서드
    def cancel_order(self):
        if len(self.order_menu) == 0:
            print("취소할 음료가 없습니다.\n")
            return
        n = -1
        while n != 0:
            n = int(input("취소할 음료 번호를 입력하세요 (0 입력 시 취소 종료): "))
            if n == 0:
                print("취소를 종료합니다.\n")
                break
            elif 1 <= n <= len(self.order_menu):
                removed_menu = self.order_menu.pop(n-1)
                removed_price = self.order_price.pop(n-1)
                print(f"{removed_menu}가 주문에서 삭제되었습니다.\n")
            else:
                print("잘못된 번호입니다. 다시 입력해주세요.\n")
    # 지불
    def pay(self):
        total = sum(self.order_price)
        print("총 결제 금액은", total, "원입니다.")

        while True:
            method = input("지불 수단을 선택하세요")

            if method == '1' or method == 'cash':
                print("직원을 호출하겠습니다: ")
                break
            elif method == '2' or method == 'card':
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("잘못된 선택입니다. 다시 결제를 시도해 주세요.\n")


    # 주문서 출력 
    def table(self):
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print('합계 금액 :', sum(self.order_price))

        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.cancel_order()    #w 주문 취소
a.pay()  # 지불
a.table()  # 주문표 출력
```
