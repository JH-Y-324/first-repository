```python
import random

class Account:

    account_count = 0
    accounts = []
    
    def __init__(self, account_holder, initial_balance):        
        self.account_holder = account_holder
        self.balance = initial_balance
        self.deposit_count = 0
        self.deposits = []
        self.withdraws = []

    def open_account(self):
        self.bank = "SC은행"
        num = str(random.randint(0, 99999999999)).zfill(11)
        self.account_number = num[:3] + '-' + num[3:5] + '-' + num[5:]

        Account.account_count += 1
        Account.accounts.append(self)

    @classmethod
    def get_account_num(self):
        print(f"Account 클래스로부터 생성된 계좌의 개수: {Account.account_count}")

    def deposit(self, money_in):
        if money_in >= 1:
            self.balance += money_in
        self.deposit_count += 1
        self.deposits.append(money_in)
        if self.deposit_count % 5 == 0:
            self.balance += 0.01*self.balance

    def withdraw(self, money_out):
        if money_out <= self.balance:
            self.balance -= money_out
            self.withdraws.append(money_out)
        else:
            print('계좌 잔고 이상의 금액을 출금할 순 없습니다.')

    def deposit_history(self):
        print(f"{self.account_holder}의 입금 내역: {self.deposits}")

    def withdraw_history(self):
        print(f"{self.account_holder}의 출금 내역: {self.withdraws}")

    def display_info(self):
        print(f"은행 이름: {self.bank}, 예금주: {self.account_holder}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

    
```

```python
a1 = Account("유정하", 990000)
a2 = Account("이선주", 1000000)
a3 = Account("윤준성", 1000000)
a1.open_account()
a2.open_account()
a3.open_account()
Account.get_account_num()
a2.deposit(50000)
a3.deposit(60000)
a3.withdraw(50000)
print()
a1.deposit_history()
a1.withdraw_history()
print()
a2.deposit_history()
a2.withdraw_history()
print()
a3.deposit_history()
a3.withdraw_history()
print()
for acount in Account.accounts:
    acount.display_info()
    
print()
for acount in Account.accounts:
    if acount.balance >= 1000000:
        acount.display_info()
```
