```python
# 사각형 넓이를 구하는 클래스 완성!
class Square:
    def __init__(self):
        self.square = int('3')

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, '3, 5'.split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        width, vertical = map(int, '7, 9'.split(','))
        area = width * vertical
        result = '평행사변형의 넓이는: ' + str(area)
        return result

    def trape(self):
        bottom_base, top_base, height = map(int, '5, 7, 9'.split(','))
        area = (bottom_base + top_base)*height*0.5
        result = '사다리꼴의 넓이는: ' + str(area)
        return result

a = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.

a.rect()
a.par()
a.trape()
```
