```python
import pandas as pd

events = pd.read_csv(f"{base_path}/events.csv")
prop1  = pd.read_csv(f"{base_path}/item_properties_part1.csv")
prop2  = pd.read_csv(f"{base_path}/item_properties_part2.csv")
cat    = pd.read_csv(f"{base_path}/category_tree.csv")

props = pd.concat([prop1, prop2], ignore_index=True)

# RetailRocket은 timestamp가 ms(밀리초)인 경우가 많아서 unit='ms'가 일반적입니다.
events["dt"] = pd.to_datetime(events["timestamp"], unit="ms")
props["dt"]  = pd.to_datetime(props["timestamp"], unit="ms")

events.head(), props.head(), cat.head()
```

```python
import numpy as np
import pandas as pd

# 1) 필요한 이벤트만
ev = events[events["event"].isin(["view", "addtocart", "transaction"])].copy()

# 2) 구매 플래그(거래ID 있는 transaction만 구매로 인정)
ev["is_buy"] = (ev["event"].eq("transaction")) & (ev["transactionid"].notna())

# 3) visitor-item 단위로 각 단계 "한 번이라도" 했는지 만들기
#    (transaction은 is_buy로)
view_flag = ev["event"].eq("view")
cart_flag = ev["event"].eq("addtocart")

user_item = (ev.assign(
                did_view=view_flag.astype(int),
                did_cart=cart_flag.astype(int),
                did_buy=ev["is_buy"].astype(int),
            )
            .groupby(["itemid", "visitorid"], as_index=False)[["did_view","did_cart","did_buy"]]
            .max()
           )

# 4) itemid별 퍼널 집계(유저 수)
funnel = (user_item
          .groupby("itemid", as_index=False)
          .agg(
              view_users=("did_view","sum"),
              cart_users=("did_cart","sum"),
              buy_users =("did_buy","sum"),
          ))

# 5) 전환율(0 나눗셈 방지)
funnel["view_to_cart"] = np.where(funnel["view_users"]>0, funnel["cart_users"]/funnel["view_users"], np.nan)
funnel["cart_to_buy"]  = np.where(funnel["cart_users"]>0, funnel["buy_users"] /funnel["cart_users"], np.nan)
funnel["view_to_buy"]  = np.where(funnel["view_users"]>0, funnel["buy_users"] /funnel["view_users"], np.nan)

# 6) 보기 좋게(예: 조회수 많은 순)
funnel = funnel.sort_values(["view_users","buy_users"], ascending=False)

funnel.head(20)
```

```python
import pandas as pd

# 구매 로그만
tx = events2[events2["event"] == "transaction"].copy()

# 분석 기준 날짜(데이터의 마지막 시점 다음날로 두는 게 일반적)
snapshot_date = events2["dt"].max() + pd.Timedelta(days=1)

# RFM 산출
rfm = (tx.groupby("visitorid")
         .agg(
             Recency=("dt", lambda x: (snapshot_date - x.max()).days),
             Frequency=("transactionid", "nunique"),   # 거래 수
             Monetary=("itemid", "count")              # 매출이 없으니 "구매 아이템 수"로 대체
         )
         .reset_index())
rfm.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = rfm[["Recency","Frequency","Monetary"]].copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

k = 4  # 예: 4개 세그먼트(원하시면 엘보우로 k 찾는 코드도 드릴게요)
km = KMeans(n_clusters=k, random_state=42, n_init="auto")
rfm["segment"] = km.fit_predict(X_scaled)

rfm["segment"].value_counts()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = rfm[["Recency","Frequency","Monetary"]].copy()

# (A) 로그 변환: Frequency/Monetary는 보통 치우침이 심함
X["Frequency"] = np.log1p(X["Frequency"])
X["Monetary"]  = np.log1p(X["Monetary"])

# (B) 상위 1% 캡(원하시면 0.5%/2% 등 조정)
for c in ["Recency","Frequency","Monetary"]:
    cap = X[c].quantile(0.99)
    X[c] = np.minimum(X[c], cap)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

km = KMeans(n_clusters=4, random_state=42, n_init="auto")
rfm["segment2"] = km.fit_predict(X_scaled)

rfm["segment2"].value_counts()
```

```python
snapshot = events["dt"].max()

tx = events[events["event"]=="transaction"].copy()

rf = tx.groupby("visitorid").agg(
    recency=("dt", lambda x: (snapshot - x.max()).days),
    frequency=("transactionid","nunique"),
    items_bought=("itemid","nunique"),
).reset_index()

rf.head()
```

```python
# =========================
# 0) 안전 체크: 필요한 DF 존재 확인
# =========================
required = ["events2", "events", "tx", "rf"]
missing = [x for x in required if x not in globals()]
if missing:
    raise NameError(f"Missing objects in memory: {missing}. 먼저 해당 DF를 생성한 뒤 실행해주세요.")

import numpy as np
import pandas as pd

# snapshot이 없다면 생성
if "snapshot" not in globals():
    snapshot = events["dt"].max()

print("snapshot:", snapshot)


# =========================
# 1) RFM + KMeans Clustering
#    - Monetary 대체: items_bought
# =========================
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

rfm = rf.copy()

# 결측/이상 방어
for c in ["recency", "frequency", "items_bought"]:
    if c not in rfm.columns:
        raise KeyError(f"rf에 {c} 컬럼이 없습니다. rf 생성 로직을 확인해주세요.")
rfm[["recency","frequency","items_bought"]] = rfm[["recency","frequency","items_bought"]].fillna(0)

# 로그 변환(분포 치우침 완화): frequency/items_bought는 +1 후 log1p
rfm_feat = rfm[["recency","frequency","items_bought"]].copy()
rfm_feat["frequency_log"] = np.log1p(rfm_feat["frequency"])
rfm_feat["items_bought_log"] = np.log1p(rfm_feat["items_bought"])

# recency는 그대로 두되, 스케일링
X = rfm_feat[["recency", "frequency_log", "items_bought_log"]].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K 선택(4~6 추천): 실루엣으로 간단 점검
sil_scores = {}
for k in [3,4,5,6,7]:
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    lbl = km.fit_predict(X_scaled)
    sil = silhouette_score(X_scaled, lbl)
    sil_scores[k] = sil

best_k = max(sil_scores, key=sil_scores.get)
print("Silhouette by k:", sil_scores)
print("Best k:", best_k)

kmeans = KMeans(n_clusters=best_k, random_state=42, n_init="auto")
rfm["cluster"] = kmeans.fit_predict(X_scaled)

# 클러스터 요약(해석용)
cluster_profile = (rfm.groupby("cluster")
                   .agg(
                       users=("visitorid","nunique"),
                       recency_mean=("recency","mean"),
                       recency_median=("recency","median"),
                       frequency_mean=("frequency","mean"),
                       frequency_median=("frequency","median"),
                       items_bought_mean=("items_bought","mean"),
                       items_bought_median=("items_bought","median"),
                   )
                   .reset_index()
                   .sort_values("users", ascending=False))

# 퍼센트
cluster_profile["user_share"] = cluster_profile["users"] / cluster_profile["users"].sum()
cluster_profile

# (선택) 페르소나 라벨 자동 부여(초안)
# 규칙 기반: recency 낮을수록 좋고, freq/items가 높을수록 좋음
def persona_row(r):
    # 상대 기준: 중앙값 기반
    # (여기선 전체 분포 중앙값을 기준으로 상/하 판단)
    return ""

med_rec = rfm["recency"].median()
med_freq = rfm["frequency"].median()
med_item = rfm["items_bought"].median()

def assign_persona(r):
    rec_good = r["recency_median"] <= med_rec
    freq_good = r["frequency_median"] >= med_freq
    item_good = r["items_bought_median"] >= med_item

    if rec_good and freq_good and item_good:
        return "VIP/핵심 충성"
    if rec_good and (freq_good or item_good):
        return "활성 우수(성장 잠재)"
    if (not rec_good) and (freq_good or item_good):
        return "휴면 위험(과거 우수)"
    if rec_good and (not freq_good) and (not item_good):
        return "신규/가벼운 활성"
    return "저활성/이탈 위험"

cluster_profile["persona"] = cluster_profile.apply(assign_persona, axis=1)
cluster_profile = cluster_profile.sort_values(["user_share","frequency_mean"], ascending=False)
cluster_profile

# rfm에 persona 붙이기
rfm = rfm.merge(cluster_profile[["cluster","persona"]], on="cluster", how="left")

print("\n[Cluster profile]")
display(cluster_profile)


# =========================
# 2) Churn 라벨 정의 + 유저 피처 집계
#    라벨: snapshot 기준 최근 30일간 어떤 이벤트도 없으면 churn=1
# =========================
from datetime import timedelta

CHURN_WINDOW_DAYS = 30
cutoff = snapshot - pd.Timedelta(days=CHURN_WINDOW_DAYS)
print("Churn cutoff:", cutoff)

# 유저별 마지막 이벤트 시점
last_evt = (events.groupby("visitorid")["dt"].max()
            .rename("last_event_dt")
            .reset_index())
last_evt["churn"] = (last_evt["last_event_dt"] < cutoff).astype(int)

# 유저별 이벤트 카운트/최근성/전환 피처
# 최근 N일(7/14/30) window counts
def window_count(df, days, event=None):
    start = snapshot - pd.Timedelta(days=days)
    x = df[df["dt"] >= start]
    if event is not None:
        x = x[x["event"] == event]
    return x.groupby("visitorid").size().rename(f"cnt_{event or 'all'}_{days}d")

# 전체 기간 이벤트 카운트
total_cnt = events.groupby("visitorid").size().rename("cnt_all_total")
view_cnt  = events[events["event"]=="view"].groupby("visitorid").size().rename("cnt_view_total")
cart_cnt  = events[events["event"]=="addtocart"].groupby("visitorid").size().rename("cnt_cart_total")
tx_cnt    = events[events["event"]=="transaction"].groupby("visitorid").size().rename("cnt_tx_total")

# 최근성(각 이벤트 마지막 발생일로부터 경과일)
def last_gap(df, event):
    s = df[df["event"]==event].groupby("visitorid")["dt"].max()
    out = (snapshot - s).dt.days.rename(f"gap_{event}")
    return out

gap_view = last_gap(events, "view")
gap_cart = last_gap(events, "addtocart")
gap_tx   = last_gap(events, "transaction")

# 최근 7/14/30일 카운트
feat_list = [
    total_cnt, view_cnt, cart_cnt, tx_cnt,
    window_count(events, 7, None),
    window_count(events, 14, None),
    window_count(events, 30, None),
    window_count(events, 7, "view"),
    window_count(events, 14, "view"),
    window_count(events, 30, "view"),
    window_count(events, 7, "addtocart"),
    window_count(events, 14, "addtocart"),
    window_count(events, 30, "addtocart"),
    window_count(events, 7, "transaction"),
    window_count(events, 14, "transaction"),
    window_count(events, 30, "transaction"),
    gap_view, gap_cart, gap_tx
]

feat = pd.concat(feat_list, axis=1).reset_index().rename(columns={"index":"visitorid"})
feat = feat.merge(last_evt[["visitorid","churn","last_event_dt"]], on="visitorid", how="left")

# 결측 처리(이벤트 경험 없는 유저는 gap이 NaN -> 큰 값으로 치환)
for c in feat.columns:
    if c.startswith("gap_"):
        feat[c] = feat[c].fillna(9999)

# 전환 여부(유저 기준)
steps = ["view","addtocart","transaction"]
user_step = (events[events["event"].isin(steps)]
             .assign(flag=1)
             .pivot_table(index="visitorid", columns="event", values="flag",
                          aggfunc="max", fill_value=0)
             .reset_index())
# 컬럼 없을 수 있으니 방어
for col in steps:
    if col not in user_step.columns:
        user_step[col] = 0

user_step["has_view"] = (user_step["view"]==1).astype(int)
user_step["has_cart"] = (user_step["addtocart"]==1).astype(int)
user_step["has_buy"]  = (user_step["transaction"]==1).astype(int)
user_step["view_to_cart_user"] = ((user_step["view"]==1) & (user_step["addtocart"]==1)).astype(int)
user_step["cart_to_buy_user"]  = ((user_step["addtocart"]==1) & (user_step["transaction"]==1)).astype(int)

feat = feat.merge(user_step[["visitorid","has_view","has_cart","has_buy","view_to_cart_user","cart_to_buy_user"]],
                  on="visitorid", how="left").fillna(0)

# RFM/cluster/persona도 합치기
feat = feat.merge(rfm[["visitorid","recency","frequency","items_bought","cluster","persona"]],
                  on="visitorid", how="left")

# 구매 경험 없는 유저의 RFM 결측 처리
for c in ["recency","frequency","items_bought","cluster"]:
    if c in feat.columns:
        feat[c] = feat[c].fillna(0)

feat["persona"] = feat["persona"].fillna("비구매/미분류")

print("\nFeature shape:", feat.shape)
feat.head()


# =========================
# 3) 베이스라인 Churn 모델 학습/평가
# =========================
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 학습에 쓸 피처 선택(문자 persona 제외)
drop_cols = ["visitorid", "last_event_dt", "persona"]
X = feat.drop(columns=[c for c in drop_cols if c in feat.columns])
y = feat["churn"].astype(int)

# 혹시 churn이 X에 남아있으면 제거
if "churn" in X.columns:
    X = X.drop(columns=["churn"])

# train/test split (층화)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# 모델1: Logistic Regression (스케일 필요 -> 간단히 StandardScaler)
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

logit = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler(with_mean=False)),  # sparse 대비
    ("model", LogisticRegression(max_iter=2000, class_weight="balanced"))
])

logit.fit(X_train, y_train)
p_logit = logit.predict_proba(X_test)[:,1]
auc_logit = roc_auc_score(y_test, p_logit)
print("Logistic AUC:", auc_logit)

# 모델2: RandomForest (해석용 중요도)
rf_clf = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("model", RandomForestClassifier(
        n_estimators=300,
        random_state=42,
        class_weight="balanced_subsample",
        max_depth=None,
        n_jobs=-1
    ))
])

rf_clf.fit(X_train, y_train)
p_rf = rf_clf.predict_proba(X_test)[:,1]
auc_rf = roc_auc_score(y_test, p_rf)
print("RandomForest AUC:", auc_rf)

# 리포트(기본 0.5 threshold)
pred_rf = (p_rf >= 0.5).astype(int)
print("\n[RandomForest classification report @0.5]")
print(classification_report(y_test, pred_rf, digits=3))

# 중요도 상위(랜덤포레스트)
rf_model = rf_clf.named_steps["model"]
importances = pd.Series(rf_model.feature_importances_, index=X_train.columns).sort_values(ascending=False)
display(importances.head(20).to_frame("importance"))


# =========================
# 4) 타겟 추출 예시: "이탈 위험 높은 활성/구매자" 리스트
# =========================
# 전체 데이터에 스코어 붙이기
feat_scored = feat.copy()
X_all = feat_scored.drop(columns=[c for c in drop_cols if c in feat_scored.columns])
if "churn" in X_all.columns:
    X_all = X_all.drop(columns=["churn"])

feat_scored["churn_score"] = rf_clf.predict_proba(X_all)[:,1]

# '현재 churn=0(최근 30일 활동 있음)' 이면서 점수 높은 유저 = "곧 이탈 위험"
at_risk = (feat_scored[(feat_scored["churn"]==0)]
           .sort_values("churn_score", ascending=False)
           .head(20))

# 같이 보면 좋은 컬럼만
cols_show = ["visitorid","churn","churn_score","persona","cluster",
             "cnt_all_7d","cnt_all_14d","cnt_all_30d",
             "cnt_view_30d","cnt_addtocart_30d","cnt_transaction_30d",
             "gap_view","gap_addtocart","gap_transaction",
             "frequency","items_bought","recency"]

display(at_risk[[c for c in cols_show if c in at_risk.columns]])
```
