# 몰랐는데 알게된 것    
where절에서 날짜를 ‘’로 감싸줘야한다

```python
select orderId, orderDate, shippedDate, status
from orders
where orderDate >= '2022-01-01'
```

where절에서 in을 쓸 때 뒤에 ()를 써야한다

```python
SELECT orderId, orderDate, status
FROM orders
WHERE status IN ('In Process' ,'Cancelled')
```

where절에서 조건을 올바르게 설정하기 위해(우선순위를 바르게 고려하게 만들기 위해) 괄호를 사용해야 할 때가 있다

```python
select orderId, shippedDate, status, price
from orders
where shippedDate is null and (status = 'In Process' or status = 'Cancelled')
```

집계합수가 포함되었을 때는 group by를 해줘야 한다    
- 틀린 코드

```python
select status as '상태', count(orderId) as '주문건수', sum(price) as '가격'
from orders
```

- 올바른 코드를 위한 원리

```python
SELECT 상태컬럼,
       집계함수1,
       집계함수2
FROM 테이블
GROUP BY 상태컬럼;
```

- 올바른 코드

```python
SELECT status AS '상태',
       COUNT(orderId) AS '주문건수',
       SUM(price) AS '가격'
FROM orders
GROUP BY status;
```

### 두 개 이상의 테이블을 join할 시, on 대신 using을 쓸 때 using 뒤에 한 개만 오더라도 괄호로 감싸줘야 한다

- 그래야 하는 주요한 이유 두 가지
    - **문법 일관성 때문에 “한 개여도 괄호를 씀”**
        
        SQL 표준(SQL:1999)에서 USING은 다음 형태여야 한다고 규정합니다:
        
        USING (column_name [, column_name]...)
        
        즉:
        
        괄호는 필수
        
        안에는 하나 또는 여러 개의 컬럼
        
        그래서 한 개일 때도 문법적으로 일관성을 유지하려고 괄호를 생략할 수 없도록 설계된 것입니다.
        
    - **SQL 파서는 USING을 “리스트”로 처리함**
        
        SQL 엔진은 USING 뒤에 나오는 것을 내부적으로 다음과 같이 인식합니다:
        
        USING ( column1, column2, column3 ... )
        
        즉, 항상 리스트로 파싱되기 때문에 괄호가 없으면 문법 구조가 깨져버립니다.

### ‘진짜로’ 복귀했는지 데이터 분석가들은 어떻게 파악하는가?

쿠폰 헌터(Coupon Hunter, 쿠폰만 쓰고 다시 사라지는 고객)인지 재진입한 고객인지 파악하고자할 때

기업들은 보통 **3단계 기준**을 사용합니다.

---

### ① 단기 행동 기준 (쿠폰 사용 직후 행동)

예:

- 쿠폰으로 1회 주문
- 쿠폰 없이 또 주문
    
    → 이 둘은 완전히 다르게 취급됩니다.
    

**쿠폰 사용 직후 7일 이내에 또 주문 있는지 여부**를 가장 중요하게 봅니다.

---

### ② 중기 행동 기준 (리텐션 유지 여부)

기업들은 보통 **14일 또는 30일 유지 판단**을 합니다.

예:

- 쿠폰 사용 → 주문
- 이후 30일 동안 1회라도 추가 주문 있음 → “재활성화 성공”
- 아무 주문 없음 → “쿠폰으로만 튀어나온 고객 = 실패”

---

### ③ 장기 행동 기준 (평균 재구매 주기와 비교)

기업은 고객군별로 “평균 주문 간격(Inter-purchase interval)”을 계산합니다.

예:

- 이 고객군의 평균 주문 간격 = 10일
- 쿠폰 사용 후 다음 주문까지 10~15일 → 정상 복귀
- 40~50일 지나도 주문 없음 → 실패

이런 방식으로 “진짜 복귀인지” 판단합니다.

---

### 이 판단은 매우 중요합니다 (실제 기업 입장)

왜냐하면 쿠폰은 **돈입니다.**

고객이 복귀하면 이익이지만, 쿠폰 먹튀라면 손해입니다.

그래서 기업은 항상 아래 두 가지만 봅니다:

- 고객 LTV (Lifetime Value)
    
    장기적으로 이 고객이 쓸 돈 예상치
    
- CAC (Customer Acquisition Cost)
    
    이 고객을 다시 데려오는 데 쓰는 비용 (쿠폰 포함)
    

결국:

> 쿠폰 비용보다 고객 lifetime value가 더 크다면 성공,
작다면 실패로 분류합니다.
