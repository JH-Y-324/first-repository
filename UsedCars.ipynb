```python
import pandas as pd
```

```python
cars = pd.read_csv('cars.csv')
brand = pd.read_csv('brand.csv')
```

데이터를 살펴봤을 때 대문자로 되어있어서 병합하려면 소문자로 바꾸는 과정이 필요하다

```python
cars['brand_name'] = cars['title'].apply(lambda x: x.split()[0].lower())
```

brand.csv의 title은 원래부터 소문자이므로 굳이 어떤 처리를 할 필요가 없다

```python
brand['brand_name'] = brand['title']
```

둘이 합친다. brand_name을 만들어줬으므로 그것을 기준으로 left join 한다.

```python
merged = pd.merge(cars, brand, on='brand_name', how='left')
```

merged 어떻게 나오는지 확인해보기

```python
merged
```

title_y 없어도 될 것 같아서 열(속성) 삭제

```python
merged = merged.drop(columns=['title_y'])
merged
```

Engine 변수에서 L을 제거하고 숫자형으로 변환

```python
merged['Engine'] = merged['Engine'].str.replace('L','').astype(float)
merged
```

Emission Class에서 뒤의 숫자만 남기기

```python
merged['Emission Class(Euro)'] = merged['Emission Class'].str.extract(r'(\d+)').astype(float)
merged
```

Service history 변수 결측치 처리 ('Unknown'으로 대체)

```python
merged['Service history'] = merged['Service history'].fillna('Unknown')
merged
```

#### 결측치 변수가 일정 개수 이상 포함된 데이터는 제거

```python
# 아까전에 Emission Class(Euro) 생성하고 drop 안 한 속성 drop
merged = merged.drop(columns='Emission Class')
```

결측치 3개 이상인 행(데이터) 확인

```python
merged['missing'] = merged.isnull().sum(axis=1)
```

```python
# 결측치 3개 이상인 행 개수
missing_morethan_2 = (merged['missing'] >= 3).sum()

# 총 행 수
total_rows = len(merged)

# 비율이 너무 크면 삭제하는 게 조금 마음에 걸려서 비율 계산
ratio_missing_3plus = (missing_morethan_2 / total_rows) * 100

print(f'{ratio_missing_3plus:.2f}%')
```

결측치 3개 이상인 행(데이터) 삭제

```python
merged = merged[merged['missing'] < 3]
```

행(데이터) 개수 줄어든 것 확인하고 missing 열도 이제 필요없으니 삭제

```python
print(len(merged))
merged['missing'].max()
merged.drop(columns='missing', inplace=True)
print(merged.columns)
```

StandardScaler로 스케일링을 하기 위해 수치형 데이터만 추출

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

numeric_data = merged.select_dtypes(include=['float64', 'int64'])
numeric_data = numeric_data.fillna(numeric_data.mean())
```

스케일링

```python
sc = StandardScaler()
sc_data = sc.fit_transform(numeric_data)
```

PCA 적용

- PCA를 적용한 결과 3개의 주성분만 남았다
- 8차원 -> 3차원으로의 축소가 이루어졌다
- 데이터의 70% 이상 정보를 유지하면서 차원을 줄인 상태이다
- 첫 번째, 두 번째, 세 번째 주성분이 각각 데이터 변동의 약 41%, 19%, 13%를 설명한다

```python
pca = PCA(n_components=0.7)
pca_data = pca.fit_transform(sc_data)

print(sc_data.shape)
print(pca_data.shape)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```
