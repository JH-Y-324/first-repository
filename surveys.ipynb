```python
import re

# df2017의 열 이름에서 HTML 태그 제거하는 함수
def remove_html_tags(text):
    return re.sub(r'<.*?>', '', text)  # HTML 태그를 빈 문자열로 대체

# df2017 열 이름에서 HTML 태그 제거
df2017.columns = [remove_html_tags(col) for col in df2017.columns]

# 2. rename_2017의 키와 df2017.columns 비교
missing_columns = [col for col in rename_2017.keys() if col not in df2017.columns]
existing_columns = [col for col in rename_2017.keys() if col in df2017.columns]

# 3. 출력
print("Missing Columns:", missing_columns)
print("Existing Columns:", existing_columns)
```

```python
df2017_filtered = df2017[existing_columns]

# 결과 확인
df2017_filtered.head()
```

```python
# 기존 열 이름을 rename_2017의 값으로 바꾸기
df2017_filtered = df2017_filtered.rename(columns=rename_2017)

# 결과 확인
df2017_filtered.head()
```

```python
# 예시: df2017_filtered를 CSV로 다운로드
df2017_filtered.to_csv('df2017_filtered.csv', index=False)

# 다운로드
files.download('df2017_filtered.csv')
```
